```python
import os
import numpy as np
import random
import matplotlib.pyplot as plt
from PIL import Image, ImageEnhance
import seaborn as sns

# TensorFlow imports
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Input, Flatten, Dropout, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications import VGG16
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.regularizers import l2

# Scikit-learn imports
from sklearn.utils import shuffle
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.model_selection import train_test_split

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)

class BrainTumorClassifier:
    def __init__(self, image_size=224, batch_size=32):
        self.image_size = image_size
        self.batch_size = batch_size
        self.class_labels = None
        self.model = None
        self.history = None
        
    def load_data(self, train_dir, test_dir, validation_split=0.2):
        """Load and organize data with optional validation split"""
        # Load training data
        train_paths = []
        train_labels = []
        
        for label in os.listdir(train_dir):
            label_dir = os.path.join(train_dir, label)
            if os.path.isdir(label_dir):
                for image in os.listdir(label_dir):
                    if image.lower().endswith(('.png', '.jpg', '.jpeg')):
                        train_paths.append(os.path.join(label_dir, image))
                        train_labels.append(label)
        
        # Load test data
        test_paths = []
        test_labels = []
        
        for label in os.listdir(test_dir):
            label_dir = os.path.join(test_dir, label)
            if os.path.isdir(label_dir):
                for image in os.listdir(label_dir):
                    if image.lower().endswith(('.png', '.jpg', '.jpeg')):
                        test_paths.append(os.path.join(label_dir, image))
                        test_labels.append(label)
        
        # Get unique class labels
        self.class_labels = sorted(list(set(train_labels)))
        print(f"Found {len(self.class_labels)} classes: {self.class_labels}")
        print(f"Training samples: {len(train_paths)}")
        print(f"Test samples: {len(test_paths)}")
        
        # Split training data into train and validation
        if validation_split > 0:
            train_paths, val_paths, train_labels, val_labels = train_test_split(
                train_paths, train_labels, 
                test_size=validation_split, 
                random_state=42,
                stratify=train_labels
            )
            return (train_paths, train_labels), (val_paths, val_labels), (test_paths, test_labels)
        else:
            return (train_paths, train_labels), None, (test_paths, test_labels)
    
    def augment_image(self, image):
        """Enhanced image augmentation"""
        if isinstance(image, np.ndarray):
            image = Image.fromarray(np.uint8(image))
        
        # Apply multiple augmentations
        # Brightness
        image = ImageEnhance.Brightness(image).enhance(random.uniform(0.7, 1.3))
        # Contrast
        image = ImageEnhance.Contrast(image).enhance(random.uniform(0.7, 1.3))
        # Sharpness (optional)
        image = ImageEnhance.Sharpness(image).enhance(random.uniform(0.8, 1.2))
        
        # Random flip
        if random.random() > 0.5:
            image = image.transpose(Image.FLIP_LEFT_RIGHT)
        
        image = np.array(image) / 255.0
        return image
    
    def preprocess_image(self, image_path, augment=False):
        """Load and preprocess single image"""
        try:
            image = load_img(image_path, target_size=(self.image_size, self.image_size))
            if augment:
                image = self.augment_image(image)
            else:
                image = np.array(image) / 255.0
            return image
        except Exception as e:
            print(f"Error loading image {image_path}: {e}")
            return None
    
    def encode_labels(self, labels):
        """Convert string labels to integers"""
        if self.class_labels is None:
            raise ValueError("Class labels not initialized. Call load_data first.")
        label_to_index = {label: idx for idx, label in enumerate(self.class_labels)}
        encoded = [label_to_index[label] for label in labels]
        return np.array(encoded)
    
    def decode_labels(self, encoded_labels):
        """Convert integer labels back to strings"""
        if self.class_labels is None:
            raise ValueError("Class labels not initialized.")
        index_to_label = {idx: label for idx, label in enumerate(self.class_labels)}
        decoded = [index_to_label[label] for label in encoded_labels]
        return decoded
    
    def data_generator(self, paths, labels, augment=False, batch_size=32):
        """Enhanced data generator with better batching"""
        paths, labels = shuffle(paths, labels, random_state=42)
        
        while True:
            for i in range(0, len(paths), batch_size):
                batch_paths = paths[i:i + batch_size]
                batch_labels = labels[i:i + batch_size]
                
                batch_images = []
                valid_labels = []
                
                for path, label in zip(batch_paths, batch_labels):
                    image = self.preprocess_image(path, augment=augment)
                    if image is not None:
                        batch_images.append(image)
                        valid_labels.append(label)
                
                if batch_images:
                    batch_images = np.array(batch_images)
                    batch_labels_encoded = self.encode_labels(valid_labels)
                    yield batch_images, batch_labels_encoded
    
    def build_model(self):
        """Build enhanced model with transfer learning"""
        # Use pre-trained VGG16 with proper initialization
        base_model = VGG16(
            input_shape=(self.image_size, self.image_size, 3),
            include_top=False,
            weights='imagenet'
        )
        
        # Freeze base model layers initially
        base_model.trainable = False
        
        # Unfreeze last few layers for fine-tuning
        for layer in base_model.layers[-4:]:
            layer.trainable = True
        
        # Build model
        self.model = Sequential([
            Input(shape=(self.image_size, self.image_size, 3)),
            base_model,
            Flatten(),
            Dropout(0.5),
            Dense(256, activation='relu', kernel_regularizer=l2(0.01)),
            Dropout(0.3),
            Dense(128, activation='relu', kernel_regularizer=l2(0.01)),
            Dropout(0.2),
            Dense(len(self.class_labels), activation='softmax')
        ])
        
        # Compile with enhanced settings
        self.model.compile(
            optimizer=Adam(learning_rate=0.0001),
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
        )
        
        print("Model built successfully!")
        print(f"Number of classes: {len(self.class_labels)}")
        self.model.summary()
    
    def train(self, train_data, val_data=None, epochs=20):
        """Train model with callbacks and validation"""
        train_paths, train_labels = train_data
        
        # Calculate steps per epoch
        steps_per_epoch = len(train_paths) // self.batch_size
        if len(train_paths) % self.batch_size != 0:
            steps_per_epoch += 1
        
        # Callbacks
        callbacks = [
            EarlyStopping(monitor='val_loss' if val_data else 'loss', 
                         patience=5, restore_best_weights=True),
            ReduceLROnPlateau(monitor='val_loss' if val_data else 'loss', 
                            factor=0.2, patience=3, min_lr=1e-7),
            ModelCheckpoint('best_model.h5', monitor='val_accuracy' if val_data else 'accuracy',
                          save_best_only=True, mode='max')
        ]
        
        # Prepare validation data
        validation_data = None
        validation_steps = None
        
        if val_data:
            val_paths, val_labels = val_data
            validation_data = self.data_generator(val_paths, val_labels, 
                                                augment=False, batch_size=self.batch_size)
            validation_steps = len(val_paths) // self.batch_size
            if len(val_paths) % self.batch_size != 0:
                validation_steps += 1
        
        # Train model
        self.history = self.model.fit(
            self.data_generator(train_paths, train_labels, augment=True, batch_size=self.batch_size),
            steps_per_epoch=steps_per_epoch,
            epochs=epochs,
            validation_data=validation_data,
            validation_steps=validation_steps,
            callbacks=callbacks,
            verbose=1
        )
    
    def evaluate(self, test_data):
        """Comprehensive model evaluation"""
        test_paths, test_labels = test_data
        
        # Preprocess all test images
        test_images = []
        valid_test_labels = []
        valid_test_paths = []
        
        for path, label in zip(test_paths, test_labels):
            image = self.preprocess_image(path, augment=False)
            if image is not None:
                test_images.append(image)
                valid_test_labels.append(label)
                valid_test_paths.append(path)
        
        test_images = np.array(test_images)
        test_labels_encoded = self.encode_labels(valid_test_labels)
        
        # Predictions
        predictions = self.model.predict(test_images, verbose=0)
        predicted_classes = np.argmax(predictions, axis=1)
        
        # Evaluation metrics
        test_loss, test_accuracy = self.model.evaluate(test_images, test_labels_encoded, verbose=0)
        
        print(f"\n{'='*50}")
        print("MODEL EVALUATION RESULTS")
        print(f"{'='*50}")
        print(f"Test Loss: {test_loss:.4f}")
        print(f"Test Accuracy: {test_accuracy:.4f}")
        
        # Classification Report
        print("\nClassification Report:")
        print(classification_report(test_labels_encoded, predicted_classes, 
                                  target_names=self.class_labels))
        
        # Confusion Matrix
        self.plot_confusion_matrix(test_labels_encoded, predicted_classes)
        
        # ROC Curves
        self.plot_roc_curves(test_labels_encoded, predictions)
        
        return test_accuracy, predictions, predicted_classes
    
    def plot_confusion_matrix(self, true_labels, predicted_labels):
        """Plot enhanced confusion matrix"""
        cm = confusion_matrix(true_labels, predicted_labels)
        
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                   xticklabels=self.class_labels, 
                   yticklabels=self.class_labels)
        plt.title('Confusion Matrix', fontsize=16, fontweight='bold')
        plt.xlabel('Predicted Labels', fontweight='bold')
        plt.ylabel('True Labels', fontweight='bold')
        plt.xticks(rotation=45)
        plt.yticks(rotation=0)
        plt.tight_layout()
        plt.show()
    
    def plot_roc_curves(self, true_labels, predictions):
        """Plot ROC curves for multi-class classification"""
        # Binarize labels for ROC curve
        true_labels_bin = label_binarize(true_labels, classes=range(len(self.class_labels)))
        
        # Compute ROC curve and ROC area for each class
        fpr, tpr, roc_auc = {}, {}, {}
        
        plt.figure(figsize=(10, 8))
        
        for i in range(len(self.class_labels)):
            fpr[i], tpr[i], _ = roc_curve(true_labels_bin[:, i], predictions[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])
            plt.plot(fpr[i], tpr[i], 
                    label=f'{self.class_labels[i]} (AUC = {roc_auc[i]:.2f})',
                    linewidth=2)
        
        # Plot diagonal line
        plt.plot([0, 1], [0, 1], 'k--', alpha=0.5, label='Random Classifier')
        
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate', fontweight='bold')
        plt.ylabel('True Positive Rate', fontweight='bold')
        plt.title('Multi-class ROC Curves', fontsize=16, fontweight='bold')
        plt.legend(loc="lower right")
        plt.grid(alpha=0.3)
        plt.tight_layout()
        plt.show()
    
    def plot_training_history(self):
        """Plot training history"""
        if self.history is None:
            print("No training history available.")
            return
        
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
        
        # Plot accuracy
        ax1.plot(self.history.history['accuracy'], label='Training Accuracy', linewidth=2)
        if 'val_accuracy' in self.history.history:
            ax1.plot(self.history.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
        ax1.set_title('Model Accuracy', fontsize=14, fontweight='bold')
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Accuracy')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # Plot loss
        ax2.plot(self.history.history['loss'], label='Training Loss', linewidth=2)
        if 'val_loss' in self.history.history:
            ax2.plot(self.history.history['val_loss'], label='Validation Loss', linewidth=2)
        ax2.set_title('Model Loss', fontsize=14, fontweight='bold')
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Loss')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
    
    def predict_single_image(self, image_path):
        """Predict class for single image"""
        try:
            image = self.preprocess_image(image_path, augment=False)
            if image is None:
                return None, None
            
            image_batch = np.expand_dims(image, axis=0)
            prediction = self.model.predict(image_batch, verbose=0)
            predicted_class_idx = np.argmax(prediction, axis=1)[0]
            confidence = np.max(prediction, axis=1)[0]
            
            predicted_class = self.class_labels[predicted_class_idx]
            
            return predicted_class, confidence
            
        except Exception as e:
            print(f"Error during prediction: {e}")
            return None, None
    
    def visualize_predictions(self, test_data, num_samples=10):
        """Visualize predictions on test samples"""
        test_paths, test_labels = test_data
        
        # Select random samples
        indices = random.sample(range(len(test_paths)), min(num_samples, len(test_paths)))
        
        fig, axes = plt.subplots(2, 5, figsize=(20, 8))
        axes = axes.ravel()
        
        for i, idx in enumerate(indices):
            if i >= 10:  # Limit to 10 images
                break
                
            image_path = test_paths[idx]
            true_label = test_labels[idx]
            
            # Make prediction
            pred_class, confidence = self.predict_single_image(image_path)
            
            # Load and display image
            img = load_img(image_path, target_size=(self.image_size, self.image_size))
            axes[i].imshow(img)
            axes[i].axis('off')
            
            # Set title with prediction results
            title_color = 'green' if pred_class == true_label else 'red'
            axes[i].set_title(f"True: {true_label}\nPred: {pred_class}\nConf: {confidence:.2f}", 
                            color=title_color, fontsize=10)
        
        plt.tight_layout()
        plt.show()
    
    def save_model(self, filepath='brain_tumor_model.h5'):
        """Save the trained model"""
        if self.model is not None:
            self.model.save(filepath)
            print(f"Model saved to {filepath}")
        else:
            print("No model to save.")
    
    def load_saved_model(self, filepath='brain_tumor_model.h5'):
        """Load a saved model"""
        try:
            self.model = load_model(filepath)
            print(f"Model loaded from {filepath}")
        except Exception as e:
            print(f"Error loading model: {e}")

# Example usage
def main():
    # Initialize classifier
    classifier = BrainTumorClassifier(image_size=224, batch_size=32)
    
    # Load data
    train_dir = '/kaggle/input/brain-tumor-mri-dataset/Training'
    test_dir = '/kaggle/input/brain-tumor-mri-dataset/Testing'
    
    train_data, val_data, test_data = classifier.load_data(train_dir, test_dir, validation_split=0.2)
    
    # Build model
    classifier.build_model()
    
    # Train model
    classifier.train(train_data, val_data, epochs=20)
    
    # Plot training history
    classifier.plot_training_history()
    
    # Evaluate model
    accuracy, predictions, pred_classes = classifier.evaluate(test_data)
    
    # Visualize predictions
    classifier.visualize_predictions(test_data, num_samples=10)
    
    # Save model
    classifier.save_model('brain_tumor_classifier.h5')
    
    # Example single prediction
    sample_image_path = '/kaggle/input/brain-tumor-mri-dataset/Testing/glioma/Te-glTr_0005.jpg'
    pred_class, confidence = classifier.predict_single_image(sample_image_path)
    print(f"\nSingle Image Prediction:")
    print(f"Image: {os.path.basename(sample_image_path)}")
    print(f"Predicted: {pred_class}")
    print(f"Confidence: {confidence:.2f}")

if __name__ == "__main__":
    main()
```
